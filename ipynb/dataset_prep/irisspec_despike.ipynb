```python
from astropy.io import fits 
from iris_prep_despike import iris_prep_despike
from tqdm.notebook import tqdm
```

```python
with fits.open("../../data/pid_1_123_aux/IRIS/iris_l2_20221024_232249_3600609177_raster_t000_r00000.fits") as hdul:
    hdul.info()
    print(repr(hdul[0].header["TDES*"]))

    new_hdul_list = []
    new_hdul_list.append(hdul[0])
    for ii in tqdm(range(1,8)):
        hdul_ = hdul[ii]
        hdul_data_despike = iris_prep_despike(hdul_.data.copy(), mode="both")
        hdul_new = fits.ImageHDU(data=hdul_data_despike, header=hdul_.header)
        new_hdul_list.append(hdul_new)
    
    new_hdul_list.append(hdul[8])
    new_hdul_list.append(hdul[9])

    new_hdul_list = fits.HDUList(new_hdul_list)

    new_hdul_list.writeto("../../data/pid_1_123_aux/IRIS/iris_l2_20221024_232249_3600609177_raster_t000_r00000_despike.fits", overwrite=True)
```
