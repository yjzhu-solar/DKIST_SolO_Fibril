Observations of different layers of the solar atmosphere within the VBI-B FOV. (a) Chromospheric fibrils in the VBI-B Hβ channel; (b) HRIEUV in the Solar Orbiter helioprojective frame; (c) HRIEUV reprojected to the Earth-based frame; (d) SDO/AIA 17.1 nm; (e) IRIS/SJI 140.0 nm; (f) VBI-R Hα; (g) VBI-B Ca II K; (h) VBI-B G-band; (i) VBI-R TiO. Panels (a) and (c-i) share the same FOV. Three ROIs and four slits are outlined in (a). The temporal evolution is available as an online movie.
